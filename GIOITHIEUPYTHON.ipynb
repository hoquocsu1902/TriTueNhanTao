```python
#13

def merge_arrays_sum(arr1, arr2):
    merged_array = []
    len1 = len(arr1)
    len2 = len(arr2)
    min_len = min(len1, len2)
    for i in range(min_len):
        merged_array.append(arr1[i] + arr2[i])
    if len1 > min_len:
        merged_array.extend(arr1[min_len:])
    elif len2 > min_len:
        merged_array.extend(arr2[min_len:])
    return merged_array
array1 = [1, 2, 3, 4, 5]
array2 = [6, 7, 8]
merged = merge_arrays_sum(array1, array2)
print(f"Mảng 1: {array1}")
print(f"Mảng 2: {array2}")
print(f"Mảng sau khi trộn: {merged}")
```

```python
import random

def create_random_matrix(m, n, min_val=1, max_val=100):
    if m <= 0 or n <= 0:
        print("Số dòng (m) và số cột (n) phải lớn hơn 0.")
        return []
    matrix = [[random.randint(min_val, max_val) for _ in range(n)] for _ in range(m)]
    return matrix
def print_matrix(matrix):
    if not matrix:
        print("Ma trận rỗng.")
        return
    for row in matrix:
        print(row)

def print_row(matrix, k):
    m = len(matrix)
    if not matrix or k < 0 or k >= m:
        print(f"Dòng {k} không hợp lệ. Ma trận có {m} dòng.")
        return
    print(f"Các phần tử thuộc dòng {k}: {matrix[k]}")

def print_column(matrix, k):
    if not matrix or not matrix[0]:
        print("Ma trận rỗng.")
        return
    n = len(matrix[0])
    if k < 0 or k >= n:
        print(f"Cột {k} không hợp lệ. Ma trận có {n} cột.")
        return
    column = [matrix[i][k] for i in range(len(matrix))]
    print(f"Các phần tử thuộc cột {k}: {column}")

def find_row_with_max_sum(matrix):
    """
    d. Tìm dòng có tổng lớn nhất.
    (Giả định 'nhất 45' là lỗi đánh máy và ý là 'lớn nhất').
    """
    if not matrix:
        return "Ma trận rỗng."
    row_sums = [sum(row) for row in matrix]
    max_sum = -float('inf')
    max_sum_row_index = -1
    for i, s in enumerate(row_sums):
        if s > max_sum:
            max_sum = s
            max_sum_row_index = i
    return f"Dòng có tổng lớn nhất là dòng {max_sum_row_index} với tổng = {max_sum}. Dòng đó là: {matrix[max_sum_row_index]}"

def find_column_with_min_product(matrix):
    """
    e. Tìm cột có tích nhỏ nhất.
    """
    if not matrix or not matrix[0]:
        return "Ma trận rỗng hoặc rỗng cột."
    m = len(matrix)
    n = len(matrix[0])
    if n == 0:
        return "Ma trận không có cột nào."

    min_product = float('inf')
    min_product_col_index = -1

    for j in range(n):
        current_product = 1
        for i in range(m):
            current_product *= matrix[i][j]

        if current_product < min_product:
            min_product = current_product
            min_product_col_index = j

    column_elements = [matrix[i][min_product_col_index] for i in range(m)]
    return f"Cột có tích nhỏ nhất là cột {min_product_col_index} với tích = {min_product}. Các phần tử của cột đó là: {column_elements}"

def print_even_rows_odd_cols(matrix):
    """
    f. Xuất ra các phần tử thuộc dòng chẵn và cột lẻ trong a.
    (Dòng/cột được đánh số từ 0).
    """
    if not matrix or not matrix[0]:
        print("Ma trận rỗng.")
        return
    m = len(matrix)
    n = len(matrix[0])
    elements = []
    for i in range(m):
        if i % 2 == 0:  # Dòng chẵn
            for j in range(n):
                if j % 2 != 0:  # Cột lẻ
                    elements.append(matrix[i][j])
    print(f"Các phần tử thuộc dòng chẵn và cột lẻ: {elements}")

def avg_even_in_odd_rows(matrix):
    """
    g. Tính trung bình cộng các phần tử chẵn thuộc dòng lẻ của a.
    (Dòng được đánh số từ 0).
    """
    if not matrix or not matrix[0]:
        return "Ma trận rỗng hoặc rỗng cột."
    m = len(matrix)
    even_elements = []
    for i in range(m):
        if i % 2 != 0:  # Dòng lẻ
            for element in matrix[i]:
                if element % 2 == 0:  # Phần tử chẵn
                    even_elements.append(element)
    if not even_elements:
        return "Không có phần tử chẵn nào thuộc dòng lẻ."
    return f"Trung bình cộng các phần tử chẵn thuộc dòng lẻ: {sum(even_elements) / len(even_elements)}"

def avg_border_elements(matrix):
    """
    h. Tính trung bình cộng các phần tử thuộc biên.
    """
    if not matrix or not matrix[0]:
        return "Ma trận rỗng hoặc rỗng cột."
    m = len(matrix)
    n = len(matrix[0])
    border_elements = []
    if m == 1 and n == 1:
        border_elements.append(matrix[0][0])
    elif m == 1:
        border_elements.extend(matrix[0])
    elif n == 1:
        for i in range(m):
            border_elements.append(matrix[i][0])
    else:
        # Hàng đầu tiên
        border_elements.extend(matrix[0])
        # Hàng cuối cùng
        border_elements.extend(matrix[m - 1])
        # Cột đầu tiên (trừ 2 phần tử đã thêm)
        for i in range(1, m - 1):
            border_elements.append(matrix[i][0])
        # Cột cuối cùng (trừ 2 phần tử đã thêm)
        for i in range(1, m - 1):
            border_elements.append(matrix[i][n - 1])

    if not border_elements:
        return "Không có phần tử biên nào (ma trận quá nhỏ)."
    return f"Trung bình cộng các phần tử thuộc biên: {sum(border_elements) / len(border_elements)}"

def avg_non_border_elements(matrix):
    """
    i. Tính trung bình cộng các phần tử không thuộc biên.
    (Giả định 'trung bình tích' là 'trung bình cộng' của các phần tử).
    """
    if not matrix or not matrix[0]:
        return "Ma trận rỗng hoặc rỗng cột."
    m = len(matrix)
    n = len(matrix[0])
    non_border_elements = []

    if m <= 2 or n <= 2:
        return "Không có phần tử không thuộc biên cho ma trận này (kích thước quá nhỏ)."

    for i in range(1, m - 1):
        for j in range(1, n - 1):
            non_border_elements.append(matrix[i][j])

    if not non_border_elements:
        return "Không có phần tử không thuộc biên."
    return f"Trung bình cộng các phần tử không thuộc biên: {sum(non_border_elements) / len(non_border_elements)}"
```

```python
import datetime

class SINHVIEN:
    def __init__(self, ma_sv, ten, nam_sinh, diem_tb):
        self.ma_sv = ma_sv[:10]  # Mã sinh viên 10 ký tự
        self.ten = ten[:20]      # Tên tối đa 20 ký tự
        self.nam_sinh = int(nam_sinh)
        self.diem_tb = float(diem_tb)

    def __str__(self):
        return f"MSV: {self.ma_sv} | Tên: {self.ten} | Năm sinh: {self.nam_sinh} | ĐTB: {self.diem_tb}"

# b. Khởi tạo mảng n sinh viên (Ví dụ mẫu)
danh_sach_sv = [
    SINHVIEN("01DH000001", "Phan Lan", 2004, 7.5),
    SINHVIEN("02DH000002", "Nguyen Van A", 2005, 4.5),
    SINHVIEN("03CD000003", "Phan Hoang", 2004, 6.0),
    SINHVIEN("04DH000004", "Tran Lan", 2003, 8.0),
    SINHVIEN("05CD000005", "Le Thi B", 2006, 5.0)
]

# c. Đếm sinh viên đủ điều kiện lên lớp (ĐTB >= 5)
def dem_sv_len_lop(ds):
    count = sum(1 for sv in ds if sv.diem_tb >= 5)
    return count

# d. Xuất sinh viên đủ 20 tuổi
def xuat_sv_20_tuoi(ds):
    current_year = datetime.datetime.now().year
    print("Danh sách sinh viên đủ 20 tuổi:")
    for sv in ds:
        if current_year - sv.nam_sinh == 20:
            print(sv)

# e. Đếm sinh viên hệ Đại học (DH ở vị trí 2,3 - index 2,3)
def dem_sv_he_dh(ds):
    count = sum(1 for sv in ds if sv.ma_sv[2:4] == "DH")
    return count

# f. Đếm sinh viên tên 'Lan'
def dem_sv_ten_lan(ds):
    # Tách từ cuối cùng của chuỗi tên để so sánh
    count = sum(1 for sv in ds if sv.ten.split()[-1].lower() == "lan")
    return count

# g. Đếm sinh viên họ 'Phan'
def dem_sv_ho_phan(ds):
    # Tách từ đầu tiên của chuỗi tên
    count = sum(1 for sv in ds if sv.ten.split()[0].lower() == "phan")
    return count

# Chạy thử nghiệm các hàm
print(f"Số SV đủ điều kiện lên lớp: {dem_sv_len_lop(danh_sach_sv)}")
xuat_sv_20_tuoi(danh_sach_sv)
print(f"Số SV hệ Đại học: {dem_sv_he_dh(danh_sach_sv)}")
print(f"Số SV tên Lan: {dem_sv_ten_lan(danh_sach_sv)}")
print(f"Số SV họ Phan: {dem_sv_ho_phan(danh_sach_sv)}")
```

```python
def merge_arrays_sum(arr1, arr2):
    merged_array = []
    len1 = len(arr1)
    len2 = len(arr2)
    min_len = min(len1, len2)

    for i in range(min_len):
        if arr1[i] < arr2[i]:
            merged_array.append(arr1[i])
        else:
            merged_array.append(arr2[i])

    if len1 > min_len:
        merged_array.extend(arr1[min_len:])
    elif len2 > min_len:
        merged_array.extend(arr2[min_len:])

    return merged_array

array1 = [1, 9, 11, 4, 5]
array2 = [6, 7, 8]
merged = merge_arrays_sum(array1, array2)
print(f"Mảng 1: {array1}")
print(f"Mảng 2: {array2}")
print(f"Mảng sau khi trộn: {merged}")
```
